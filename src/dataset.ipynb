# Dataset Overview

```python
import warnings

warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use("fivethirtyeight")
FIGSIZE = (16, 9)
```

```python
path = "../data/responses_cleaned.csv"
df = pd.read_csv(path, header=0)
```

```python
groups = {
"Music preferences": 19,
"Movie preferences": 12,
"Hobbies & interests": 32,
"Phobias": 10,
"Health habits": 3,
"Personality traits, views on life, & opinions": 57,
"Spending habits": 7,
"Demographics": 10
}
```

```python
fig = plt.figure(figsize=(5,5))
plt.title("Survey groups", fontsize=16)

# Create a pieplot
plt.pie(
    groups.values(),
    labels=groups.keys(),
    colors=sns.color_palette("mako"),
    textprops={'fontsize': 10}
)
# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color="white")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
plt.title("Gender Distribution", fontsize=14)

male, female = len(df[df["Gender"] == "male"]), len(df[df["Gender"] == "female"])

# Create a pieplot
plt.pie(
    [male, female],
    labels=[f"Male {100 * male / (len(df)):.0f}%", f"Female {100 * female / (len(df)):.0f}%"],
    colors=["black", "darkcyan"],
    textprops={'fontsize': 10}
)
# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color="white")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize=(FIGSIZE))
plt.title("Age distribution", fontsize=24)
sns.countplot(data=df, x="Age", color="darkcyan")
plt.show()
```

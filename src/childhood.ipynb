# Childhood Impact
Analyze if there are differences between the participants based on where they lived most of their childhood: rural and urban area.

```python
import warnings

warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.feature_selection import r_regression

plt.style.use("fivethirtyeight")
FIGSIZE = (16, 6)
```

```python
path = "../data/responses_cleaned.csv"
df = pd.read_csv(path, header=0)
```

## Correlation

```python
cols = ["Music","Slow songs or fast songs","Dance","Folk","Country",
        "Classical music","Musical","Pop","Rock","Metal or Hardrock",
        "Punk","Hiphop, Rap","Reggae, Ska","Swing, Jazz","Rock n roll",
        "Alternative","Latino","Techno, Trance","Opera"]

plt.figure(figsize=FIGSIZE)
plt.title("Music preference")
corr_coeff = r_regression(df[cols], pd.get_dummies(df["Village - town"])["village"])
# Sort the cols based on corr_coeff
cols = [c for c, coef in sorted(zip(cols, corr_coeff), key=lambda pair: pair[1])]
corr_coeff = sorted(corr_coeff)
plt.bar(cols, corr_coeff)
plt.xticks(cols, rotation=90)
plt.yticks(np.linspace(-0.2, 0.2, 11))
plt.show()
```
